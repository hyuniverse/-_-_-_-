```python
rch=rd.rdarray(demarr,no_data=-9999)
```

```python
slope=rd.TerrainAttribute(rch,attrib='slope_degrees')
```

```python
slope
```

```python
rd.rdShow(slope,axes=False,cmap='Greens',figsize=(16,12))
```

```python
def np_to_raster(array,savename,baseTIF):
    image=rasterio.open('./data/%s.tif'%(savename),'w',driver='Gtiff', width=baseTIF.width, height=baseTIF.height,count=1, crs=baseTIF.crs, transform=baseTIF.transform,dtype='float64')
    image.write(array,indexes=1)
    image.close()
```

```python
np_to_raster(slope,'icn_slopeDg',dem)
```

```python
slope_rst=rasterio.open('./data/icn_slopeDg.tif')
fig,ax=plt.subplots(figsize=(10,10))
show(slope_rst,cmap='Greens')
```

## 경사도 점수 래스터 생성

```python
slope_sc=np.zeros_like(demarr)
```

```python
for y in range(len(slope)):
    for x in range(len(slope[0])):
        if slope[y][x]>60:
            slope_sc[y][x]=0
        elif slope[y][x]>35:
            slope_sc[y][x]=2
        else:
            slope_sc[y][x]=3
slope_sc
```

```python
np_to_raster(slope_sc,'icn_slope_sc',dem)
```

```python
slope_sc_rst=rasterio.open('./data/icn_slope_sc.tif')
fig,ax=plt.subplots(figsize=(10,10))
show(slope_sc_rst,cmap='Greens')
```

## 자전거 도로

```python
def rasterize(data, vectorSrc, field, outFile):
    dataSrc = gdal.Open(data)
    import ogr
    shp = ogr.Open(vectorSrc)

    lyr = shp.GetLayer()

    driver = gdal.GetDriverByName('GTiff')
    dst_ds = driver.Create(
        outFile,
        dataSrc.RasterXSize,
        dataSrc.RasterYSize,
        1,
        gdal.GDT_UInt16)
    dst_ds.SetGeoTransform(dataSrc.GetGeoTransform())
    dst_ds.SetProjection(dataSrc.GetProjection())
    if field is None:
        gdal.RasterizeLayer(dst_ds, [1], lyr, None)
    else:
        OPTIONS = ['ATTRIBUTE=' + field]
        gdal.RasterizeLayer(dst_ds, [1], lyr, None, options=OPTIONS)

    data, dst_ds, shp, lyr = None, None, None, None
    return outFile 
```
