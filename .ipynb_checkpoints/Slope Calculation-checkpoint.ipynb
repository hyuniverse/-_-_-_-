```python
def np_to_raster(array,savename,baseTIF):
    image=rasterio.open('./data/%s.tif'%(savename),'w',driver='Gtiff', width=baseTIF.width, height=baseTIF.height,count=1, crs=baseTIF.crs, transform=baseTIF.transform,dtype='float64')
    image.write(array,indexes=1)
    image.close()
```

```python
def count_usage(usage_data,date):
    start=[]
    end=[]
    for poi in poigeo['대여소_ID']:
        start.append(len(usage_data.loc[usage_data['시작_대여소_ID']==poi]))
        end.append(len(usage_data.loc[usage_data['종료_대여소_ID']==poi]))

    poigeo['%s대여'%(date)]=gpd.GeoDataFrame(start)
    poigeo['%s반납'%(date)]=gpd.GeoDataFrame(end)
```

### 대여건수가 적은 곳과 많은 곳의 건물분포 평균 비교
대여소 포인트를 기준으로 100m 버퍼한 영역 내에서 카운트한 건물 수를 종류별로 평균 내어 비교함. 
아래는 차이가 어느정도 난다고 판단한 건물군을 정리해놓은 것. 


|건물코드|건물종류|하위100|상위100|
|------|------|-----|----|
|BDS001|주택|18.09|10.66|
|BDS002|근린생활시설|13.04|22.69|
|BDS999|기타시설|3.40|4.72|
|BDS008|교육연구시설|1.61|0.41|
|BDS012|업무시설|0.42|0.95|
|BDS013|숙박시설|0.19|0.54|


분석결과: 
주어진 데이터로만 본다면
주택이 있는 곳, 근린생활시설이 다수 분포한 곳은 기본적으로 수요가 있는 편임. 
그러나 근린생활시설 비율이 높은 곳에서의 수요가 더 컸음. 
업무시설의 경우 예상보다 수요가 적었음. 
원인으로는 애초에 건물 비율 자체에 큰 차이가 있을 것으로 생각해, 전체 건물 비율과 대여건수를 한 눈에 볼 수 있도록 그래프를 그려봄. 

```python
from collections import Counter
```

```python
bd_serv=Counter(bd['SERV'])
bd_serv
```

```python
bd_serv=dict(sorted(bd_serv.items()))
bd_serv
```

```python
fig = plt.figure(figsize=(30,20)) ## Figure 생성 
ax1 = fig.add_subplot() ## axes 생성
 

ax1.bar(bd_serv.keys(),bd_serv.values(),alpha=0.5) ## 바차트 출력
 

ax2 = ax1.twinx() ## 새로운 axis 생성
ax2.plot(bd_vs.index, bd_vs['top_mean'], color='red', linestyle='--', marker='o') ## 선 그래프 
ax2.plot(bd_vs.index, bd_vs['bot_mean'], color='black', linestyle='--', marker='o')
ax2.tick_params(axis='y', labelcolor='red') ## 눈금 라벨 색상 지정

 
plt.title('대여건수 정도에 따른 건물 분포 비교', fontsize=20)
plt.show()
```

예상대로 주택과 근린생활시설의 수 자체가 많기 때문에 분포비율이 높은 것이었음. \
따라서 근린생활시설 및 주택가가 유달리 수요가 높은 것이 아닌, 절대량이 많은 것에 의한 결과임을 알 수 있음. \
때문에 용도별 건물이 대여건수 차이에 미치는 영향을 분석함에 있어서 절대량 보다는, 차이에 중점을 두고 보아야할 것으로 판단함. \
\
또한 용도별 건물이 미치는 영향력은 수요대비 건물 수를 고려해야할 것으로 판단함.\
가령 업무시설과 숙박시설을 비교해보았을 때, 두 곳 모두 상위100군집에서 더 높은 비율로 나타난 용도건물이지만, 수요와 건물 수를 고려해보았을 때 업무시설의 경우 더 적은 절대량을 갖고 있으나 수요에 있어서 더 큰 값을 가지고 있기때문에 숙박시설보다 수요에 미치는 영향이 크다고 볼 수 있음.\
따라서 하위100군집과 상위100군집 사이에 어느정도 차이가 있다고 본 용도건물만 정리한 결과는 다음과 같음.

|건물코드|건물종류|하위100|상위100|
|------|------|-----|----|
|BDS001|주택|18.09|10.66|
|BDS002|근린생활시설|13.04|22.69|
|BDS999|기타시설|3.40|4.72|
|BDS008|교육연구시설|1.61|0.41|
